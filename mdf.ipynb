```python
from pathlib import Path
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# File names
datafile = 'dr17_cut_data.csv'
data_dir = '/mnt/c/Users/dubay.11/OneDrive - The Ohio State University/Documents/Docs/First Year Project/Data'

# Plot settings
plt.rcParams.update({
    'xtick.direction': 'in',
    'xtick.top': True,
    'ytick.direction': 'in',
    'ytick.right': True,
    'figure.dpi': 140,
    'figure.figsize': (4, 4),
    'font.family': 'STIXgeneral',
    'mathtext.fontset': 'cm',
})
```

```python
data = pd.read_csv(Path(data_dir) / datafile)
data
```

```python
abszlim = (0, 0.5) # kpc
galrlim = (7, 9)
galphilim = (-np.pi, np.pi)

zbound = (np.abs(data['ASTRONN_GALZ']) >= abszlim[0]) & (np.abs(data['ASTRONN_GALZ']) < abszlim[1])
rbound = (np.abs(data['ASTRONN_GALR']) >= galrlim[0]) & (np.abs(data['ASTRONN_GALR']) < galrlim[1])
phibound = (np.abs(data['ASTRONN_GALPHI']) >= galphilim[0]) & (np.abs(data['ASTRONN_GALPHI']) < galphilim[1])
region = zbound & rbound & phibound

data[region]
```

```python
mdf = data[region]['ASTRONN_FE_H']
dfe = 0.05
bins = np.arange(-2.1, 0.6+dfe, dfe)
fig, ax = plt.subplots()
ax.hist(mdf, bins=bins)
ax.set_yscale('log')
ax.set_ylabel('dN/d[Fe/H]')
ax.set_xlabel('[Fe/H]')
plt.show()
```

```python
odf = data[region]['ASTRONN_O_FE']
do = 0.05
bins = np.arange(-0.6, 0.6+dfe, dfe)
fig, ax = plt.subplots()
ax.hist(odf, bins=bins)
ax.set_yscale('log')
ax.set_ylabel('dN/d[O/Fe]')
ax.set_xlabel('[O/Fe]')
plt.show()
```
